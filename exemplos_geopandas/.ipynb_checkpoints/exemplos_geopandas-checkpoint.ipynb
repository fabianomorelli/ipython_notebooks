# Exemplo de cruzamento espacial utilizando dois shapefiles
## Filtra os pontos que estão contidos no polígono

```python
# Utiliza um comando magico para rodar tudo dentro de um jupyter notebook
# desnecessário para um script
%matplotlib inline
```

```python
# Declara a importacao das bibliotecas a serem utilizadas
import geopandas as gpd
```

```python
# carrega o shapefile que está no mesmo diretorio do programa atual
# este arquivo eh o limite politico do brasil

brasil = gpd.read_file("./brasil_world_countries.shp")
```

```python
# carrega o shapefile que está no mesmo diretorio do programa atual
# este arquivo eh um shapefile do bdqueimadas com focos de toda America do Sul

focos_ams = gpd.read_file("./Focos.2018-05-01.2018-05-14.shp")
```

```python
# Apresenta o shapefile na forma de um mapa
# isto eh pra demonstrar como fazer uma figura
brasil.plot(figsize=(6,12));
```

```python
# Apresenta a descrição do arquivo que agora eh um geodataframe
# isso eh pra demonstrar como ver a descricao do arquivo
brasil.info()
```

```python
# Este comando serve para mostrar as primeiras linhas do dataframe
# neste caso existe apenas uma linha
brasil.head()
```

```python
# este comando serve pra mostrar o tamanho da matriz ou dataframe
# neste caso possui uma linha e quatro colunas
# ou seja, um registro com quatro atributos
brasil.shape
```

```python
# note que a quantidade de focos e o numero de atributos do arquivo que foi lido.
focos_ams.shape
```

```python
# agora efetivamente o cruzamento espacial que ira filtrar apenas os focos contidos no poligono
focos_brasil = gpd.sjoin(focos_ams, brasil, how='inner', op='within')

# voce pode utilizar outros metodos no parametro how para definir o retorno
# por exemplo, se utilizar how='left' vai retornar todos os focos e colocar atributos
# do shapefile brasil apenas naqueles que estiverem contidos no polígono
```

```python
focos_brasil.head(5)
# note que até a coluna FRP são atributos que estavam nos focos
# depois sao apresentados todos os atributos que estavam no shapefile Brasil
```

```python
tabela_focos_estados = focos_brasil.Estado.value_counts()
```

```python
tabela_focos_estados.sort_index()
```

```python
# exporta os dados para um arquivo csv
tabela_focos_estados.to_csv("./tabela_focos_estados.csv")
```

```python
# exporta todos os dados do dataframe para um arquivo shapefile
focos_brasil.to_file(driver = 'ESRI Shapefile', filename= "./focos_brasil.shp")
```

# Exemplo que utiliza um arquivo CSV como entrada

```python
# Declara a importacao das bibliotecas a serem utilizadas
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
```

```python
firms_file = "./fire_nrt_M6_10073.csv"
```

```python
# esta declaracao eh opcional mais serve pra mostrar onde podemos forcar que um atributo seja 
# de um tipo especifico
tipo_dados = {'acq_date': str, 'acq_time': str, 'latitude': float, 'longitude': float}
```

```python
firms_modis_c6 = pd.read_csv(firms_file, dtype=tipo_dados)
```

```python
firms_modis_c6.head()
```

```python
list(firms_modis_c6.acq_date.str)
```

```python
firms_modis_c6.info()
```

```python
# Prepara um novo atributo do tipo geometry utilizando a funcao Point do shapely para depois criar um geopandas
# esta variavel row eh soh pra ajudar a entender que ele vai pegar pra cada linha o atributo entre chaves

geometry = firms_modis_c6.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)

# cria uma variavel com referencia ao sistema de projecao cartografica
crs = {'init': 'epsg:4326'}


# cria um geodataframe com o dataframe anteriro
# neste caso usando o mesmo nome mais pode ser outro qualquer
firms_modis_c6 = gpd.GeoDataFrame(firms_modis_c6, crs=crs, geometry=geometry)

```

```python
type(firms_modis_c6)
```

```python
# Verificar os atributos do dado de entrada
firms_modis_c6.info()
```

```python
# Apresenta o mapa dos pontos
firms_modis_c6.plot(figsize=(24,12));
```

```python
firms_modis_c6_brasil = gpd.sjoin(firms_modis_c6, brasil, how='inner', op='within')
```

```python
firms_modis_c6_brasil.plot(figsize=(24,12));
```
